```python

import requests

# URL del endpoint de la API
url = 'https://pokeapi.co/api/v2/pokemon/'

# Realizar la petición GET
response = requests.get(url)

# Verificar si la petición fue exitosa (código de estado 200)
if response.status_code == 200:
    # Guardar los resultados en una variable
    resultados = response.json()
else:
    print("Error al realizar la petición:", response.status_code)

# Usar la variable 'resultados' según sea necesario
#print("Conteo total de Pokémon:", resultados['count'])
#print("Próxima página de resultados:", resultados['next'])
#print("Pokémon encontrados:")
for pokemon in resultados['results']:
    print(pokemon['name'])
```

```python
pokemones_con_estadisticas = {}

# Iterar sobre cada Pokémon en los resultados
for pokemon in resultados['results']:
    nombre = pokemon['name']
    url_pokemon = pokemon['url']
    
    # Hacer una solicitud GET para obtener los detalles del Pokémon
    response_pokemon = requests.get(url_pokemon)
    
    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response_pokemon.status_code == 200:
        # Guardar los detalles del Pokémon en una variable
        detalles_pokemon = response_pokemon.json()
        
        # Extraer las estadísticas relevantes del Pokémon
        stats = detalles_pokemon['stats']
        estadisticas = {}
        for stat in stats:
            stat_nombre = stat['stat']['name']
            stat_valor = stat['base_stat']
            estadisticas[stat_nombre] = stat_valor
       
        
        # Agregar el Pokémon y sus estadísticas al diccionario
        pokemones_con_estadisticas[nombre] = estadisticas
    else:
        print("Error al obtener detalles del Pokémon:", response_pokemon.status_code)

# Imprimir el diccionario con los Pokémon y sus estadísticas
for nombre, estadisticas in pokemones_con_estadisticas.items():
    print(nombre, ":", estadisticas)
```

```python

def combate(pokemon1, pokemon2):
    print("¡Comienza el combate!")
    
    # Determinar aleatoriamente el orden de los turnos
    orden_turnos = [pokemon1, pokemon2] 
    
    while True:
        for atacante, defensor in [(orden_turnos[0], orden_turnos[1]), (orden_turnos[1], orden_turnos[0])]:
            # Calcular el daño del ataque
            dano =  atacante['ataque'] - defensor['defensa']
            # Reducir la salud del Pokémon defensor
            defensor['salud'] -= dano

            print(f"{atacante['nombre']} ataca a {defensor['nombre']} y hace {dano} puntos de daño.")
            # Verificar si el Pokémon defensor ha sido derrotado
            if defensor['salud'] <= 0:
                print(f"{defensor['nombre']} ha sido derrotado!")
                print(f"{atacante['nombre']} es el ganador!")
                return

# Ejemplo de uso
pokemon1 = pokemon1
pokemon2 = pokemon2

combate(pokemon1, pokemon2)
```
